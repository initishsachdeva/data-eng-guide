```python
import pandas as pd
```

```python
df = pd.read_csv("09_pivot_table.csv")
df
```

# Change the above data set to Pivot

```python
df.pivot(index = "date", columns= "city")
```

```python
df.pivot(index = "date", columns= "city", values = "humidity")
```

# Pivot Table - to summarize and aggregate tabular data

```python
df.pivot_table(index = "city", columns= "date")
```

```python
df.pivot_table(index = "city", columns= "date", aggfunc= "count")
```
