```python
import pandas as pd
import numpy as np
```

# if you have a data with missing values

```python
df = pd.read_csv("05_missing_values.csv", parse_dates= ["day"])
df
```

```python
df.set_index('day', inplace = True)
df
```

# replace  all NAN values with some other value using fillna method

```python
new_df = df.fillna(0)
new_df
```

# sometime 0 is not an exactly or correct value to be replaced for NAN values. Therefore if we specifiy diff values for diff cols by using dictionary`

```python
new_df = df.fillna({
    'temp': 0,
    'windspeed': 0,
    'event': 'no event'
}
)
new_df
```

# if we want to carry values from previous row

```python
new_df = df.fillna(method = "ffill")
new_df
```

# if we want to carry current value to previous empty row, use bfill

```python
new_df = df.fillna(method = "bfill")
new_df
```

# if you want to copy or move values horizontally to empty row, then use axis

```python
new_df = df.fillna(method = "bfill", axis="columns")
new_df
```

# if you want copy the values only once , then use limit property

```python
new_df = df.fillna(method = "ffill", limit = 1)
new_df
```

# if you want to get close values bw 2 diff cells like temp A = 20, B= NAN, C= 25 the B might be having value bw 20-25 ,use interpolate method. By default it will show result in linear

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

# if you want to drop na values, use dropna.By defaul it will drop all the rows which has Na values.

```python
new_df = df.dropna()
new_df
```

# but if you want to preserve some rows which has atleast 1 data values

```python
new_df = df.dropna(thresh=2)
new_df
```

# if you want to fill missing values for eg in dates if you have some missing dates 

```python
dt = pd.date_range("01-02-2024","01-11-2024")
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(idx)
new_df
```

# How to use replace methods to add any new or update any existing values

```python
new_df = df.replace([-23, -5], np.nan)
new_df
```

# replace values based on sp[ecific cols]

```python
new_df = df.replace({
    'temp': -23,
    'windspeed': -5,
    'event': 'snow'
     }, np.nan)
new_df
```
