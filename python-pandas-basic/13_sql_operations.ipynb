```python
import pandas as pd
import sqlalchemy
```

# engine = sqlalchemy.create_engine('mysql+pymysql://username:pwd@localhost:3306/db_name')

```python
engine = sqlalchemy.create_engine('mysql+pymysql://:@localhost:3306/data_engg')
```

# To read or select rows from the table
# use columns arguments if you want to select specific columns from the table

```python
df = pd.read_sql_table("categories", engine, columns=['category_id', 'category_name'])
df
```

```python
query = '''
SELECT 
    o.*, oi.order_item_product_id, oi.order_item_subtotal
FROM
    orders AS o
        JOIN
    order_items AS oi ON o.order_id = oi.order_item_order_id;

'''

df = pd.read_sql_query(query, engine)
df
```

# if the records are huge, use the 'chunksize' argument to view the records in batches

```python
query = '''
SELECT 
    o.*, oi.order_item_product_id, oi.order_item_subtotal
FROM
    orders AS o
        JOIN
    order_items AS oi ON o.order_id = oi.order_item_order_id;

'''

df = pd.read_sql_query(query, engine,chunksize=100 )
df
```

# Copy data from csv file and write into database table

```python
df = pd.read_csv("13_write_to_db.csv")
df.to_sql(
    name = "categories",    # db table name
    con = engine,           # connection created with sql
    index = False,          # mark index = False, if you don't want datasets index
    if_exists = "append"    # use this to update current table  
    )
```

# instead of using read_sql_table and read_sql_query , you can use read_sql() method and pass the same details you pass in other 2 methods

```python
df = pd.read_sql("categories", engine, columns=['category_id', 'category_name'])
df

```
