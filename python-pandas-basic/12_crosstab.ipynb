```python
import pandas as pd
import numpy as np
df = pd.read_csv("12_crosstab.csv")
df
```

# Crosstab - In statistics, a contingency table (also known as a cross tabulation or crosstab) is a type of table in a matrix format that displays the multivariate frequency distribution of the variables. 

```python
df1 = pd.crosstab(df.Nationality, df.Handedness)
df1
```

```python
df1 = pd.crosstab(df.Sex, df.Handedness)
df1
```

# We can also get the totals by using margins = True

```python
df1 = pd.crosstab(df.Nationality, df.Handedness, margins = True)
df1
```

```python
df1 = pd.crosstab(df.Nationality, [df.Handedness, df.Nationality])
df1
```

```python
df1 = pd.crosstab(df.Sex, [df.Handedness], normalize='index')
df1
```

# find avergae age 

```python
df1 = pd.crosstab(df.Sex, [df.Handedness], values= df.Age, aggfunc=np.average)
df1
```
